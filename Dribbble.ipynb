Get Likes, views, category (pro, team, boosted), creator name for each sub tab and major tab

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page = []
category = []
likes = []
views = []
names = []
badges = []

def scrape(url, likes, views, names, badges, page, category):
    html_text = requests.get(url).text
    soup = BeautifulSoup(html_text, 'lxml')

    likes_bs4 = soup.find_all('span', class_ = "js-shot-likes-count color-deep-blue-sea-light-20 font-weight-500")
    views_bs4 = soup.find_all('span', class_ = "js-shot-views-count color-deep-blue-sea-light-20 font-weight-500")
    names_bs4 = soup.find_all('span', class_ = "display-name")
    userinfo_bs4 = soup.find_all('div', class_ = "user-information")

    for i in likes_bs4:
        x = i.get_text()
        x = x.replace(' ','')
        x = x.replace('\n','')
        likes.append(x)

    for i in views_bs4:
        x = i.get_text()
        x = x.replace(' ','')
        x = x.replace('\n','')
        views.append(x)

    for i in names_bs4:
        x = i.get_text()
        x = x.replace(' ','')
        x = x.replace('\n','')
        names.append(x)
        if 'popular' in url:
            page.append('Popular')
        else:
            page.append('New & Noteworthy')
            
        if 'animation' in url:
            category.append('Animation')
        elif 'branding' in url:
            category.append('Branding')
        elif 'illustration' in url:
            category.append('Illustration')
        elif 'mobile' in url:
            category.append('Mobile')
        elif 'print' in url:
            category.append('Print')
        elif 'product-design' in url:
            category.append('Product Design')
        elif 'typography' in url:
            category.append('Typography')
        elif 'web-design' in url:
            category.append('Web Design')
        else:
            category.append('ERROR')
            
    for i in userinfo_bs4:
        x = i.get_text()
        x = x.replace(' ','')
        x = x.replace('\n','')
        if 'Team' in x:
            badges.append('Team')
        elif 'Pro' in x:
            badges.append('Pro')
        else:
            badges.append('None')
            
        
    return likes, views, names, badges, page, category
```

```python
likes, views, names, badges, page, category = scrape('https://dribbble.com/shots/popular/animation', likes, views, names, badges, page, category)
likes, views, names, badges, page, category = scrape('https://dribbble.com/shots/popular/branding', likes, views, names, badges, page, category)
likes, views, names, badges, page, category = scrape('https://dribbble.com/shots/popular/illustration', likes, views, names, badges, page, category)
likes, views, names, badges, page, category = scrape('https://dribbble.com/shots/popular/mobile', likes, views, names, badges, page, category)
likes, views, names, badges, page, category = scrape('https://dribbble.com/shots/popular/print', likes, views, names, badges, page, category)
likes, views, names, badges, page, category = scrape('https://dribbble.com/shots/popular/product-design', likes, views, names, badges, page, category)
likes, views, names, badges, page, category = scrape('https://dribbble.com/shots/popular/typography', likes, views, names, badges, page, category)
likes, views, names, badges, page, category = scrape('https://dribbble.com/shots/popular/web-design', likes, views, names, badges, page, category)
likes, views, names, badges, page, category = scrape('https://dribbble.com/shots/recent/animation', likes, views, names, badges, page, category)
likes, views, names, badges, page, category = scrape('https://dribbble.com/shots/recent/branding', likes, views, names, badges, page, category)
likes, views, names, badges, page, category = scrape('https://dribbble.com/shots/recent/illustration', likes, views, names, badges, page, category)
likes, views, names, badges, page, category = scrape('https://dribbble.com/shots/recent/mobile', likes, views, names, badges, page, category)
likes, views, names, badges, page, category = scrape('https://dribbble.com/shots/recent/print', likes, views, names, badges, page, category)
likes, views, names, badges, page, category = scrape('https://dribbble.com/shots/recent/product-design', likes, views, names, badges, page, category)
likes, views, names, badges, page, category = scrape('https://dribbble.com/shots/recent/typography', likes, views, names, badges, page, category)
likes, views, names, badges, page, category = scrape('https://dribbble.com/shots/recent/web-design', likes, views, names, badges, page, category) 
```

```python
data = pd.DataFrame(columns=['Page', 'Category', 'Badge', 'Name', 'Likes', 'Views'])
data['Page'] = page
data['Category'] = category
data['Badge'] = badges
data['Name'] = names
data['Likes'] = likes
data['Views'] = views

data.to_csv(r'C:\Users\amind\Juptyer lab Projects\Dribbble\Data.csv', index=False)  
```
